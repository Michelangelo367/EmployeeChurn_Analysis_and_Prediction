### Import Packages

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import pylab
import pylab as pl
```

```python
dataset = pd.read_csv("Employeechurn.csv")
```

### Histogram function definition

```python
def histogram(data, x_label, y_label, title):
    """Histogram function definition consist of all parameters to define the characterstics of histograms """
    _, ax = plt.subplots()
    ax.hist(data, color = '#539caf',bins= 25,density=False,histtype='bar',align='mid', orientation='vertical',edgecolor='black')
    ax.set_ylabel(y_label)
    ax.set_xlabel(x_label)
    ax.set_title(title)
```

### Age vs Attrition

```python
a = histogram(data = dataset['Age']
           , x_label = 'Age'
           , y_label = 'Attrition'
           , title = 'Age vs Attrition')
```

```python
a1 = dataset['Age'].hist(by=dataset['Attrition'],bins= 25,density=False,histtype='bar',align='mid', orientation='vertical',edgecolor='black')
for ax in a1.flatten():
    ax.set_xlabel("Age")
    ax.set_ylabel("Attrition")
```

```python
#Function
def jobrole(data, x_label, y_label, title,a,c):
    _, ax = plt.subplots()
    ax.hist(data, color = '#539caf',bins= a,density=False,histtype='bar',align='mid', orientation='vertical',edgecolor='black')
    ax.set_ylabel(y_label)
    ax.set_xlabel(x_label)
    ax.set_xticklabels(c,rotation=90)
    ax.set_title(title)
```

### JobRole vs Attrition

```python
v = jobrole(data = dataset['JobRole']
           , x_label = 'JobRole'
           , y_label = 'Attrition'
           , title = 'JobRole vs Attrition'
           , a = 9
           , c = dataset['JobRole'])
v1 = dataset['JobRole'].hist(by=dataset['Attrition'],bins= 9,density=False,histtype='bar',align='mid', orientation='vertical',edgecolor='black')
for ax in v1.flatten():
    ax.set_xlabel("JobRole")
    ax.set_ylabel("Attrition")
```
